```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Customer Churn.csv') #import 
df.head() #view first few rows of a dataframe (default - 5 rows)
```

```python
df.info() #inspection
```

#replacing blank spaces with 0 as tenure is 0 and total charges are not recorded.

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
"""
Converting 0 and 1 as yes and no respectively, to make it easier to understand
code method 01

def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
"""

#simpler code
#code method 02
df['SeniorCitizen'] = df['SeniorCitizen'].apply(lambda value: "yes" if value == 1 else "no")
```

```python
df.head()
```

#creates a barplot where the height of each bar represents the count of observations in each category

```python
# Define custom colors for the countplot
custom_colors = ['#FFD700', '#8A2BE2']  # Gold and Blue Violet

# Create the countplot
ax = sns.countplot(x='Churn', data=df, hue='Churn', palette=custom_colors)

# Add bar labels
ax.bar_label(ax.containers[0])

# Set the title of the plot
plt.title("Count of Customers by Churn")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(3, 4))  # Resize the pie chart

# Group the data by 'Churn' and aggregate counts
gb = df.groupby("Churn").agg({'Churn': "count"})

# Define custom colors for the pie chart
custom_colors = ['#FFD700', '#8A2BE2']  # Gold and Blue Violet

# Create the pie chart
plt.pie(
    gb['Churn'], 
    labels=gb.index, 
    autopct="%1.2f%%", 
    colors=custom_colors
)

# Add a title
plt.title("Percentage of Churned Customers", fontsize=12)
plt.show()
```

#From the pie chart, we observe that 26.54% of our customers have churned. Let's now explore the reasons behind this customer churn.

```python
plt.figure(figsize=(3, 3))  # Resize 

# Add a title
plt.title("Churned Customers by Gender", fontsize = 10)

sns.countplot(x = "gender", data = df, hue='Churn', palette=custom_colors)
plt.show()
```

```python
plt.figure(figsize=(3, 3))  # Resize 

# Add a title
plt.title("Churned Customers by Senior Citizen", fontsize = 10)

sns.countplot(x = "SeniorCitizen", data = df, hue='Churn', palette=custom_colors)
plt.show()
```

```python
plt.figure(figsize=(3, 3))  # Resize 

# Add a title
plt.title("Count of Senior Citizen Customers", fontsize = 10)


ax.bar_label(ax.containers[0])

sns.countplot(x = "SeniorCitizen", data = df, palette=custom_colors)
plt.show()
```

```python
# Calculate the count of churned and non-churned customers for each SeniorCitizen category
churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate the percentage
churn_percentage = churn_counts.div(churn_counts.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
plt.figure(figsize=(3, 4))  # Resize the figure
ax = churn_percentage.plot(kind='bar', stacked=True, color=custom_colors, figsize=(6, 4))

# Add labels for percentage on the bars
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x, y = p.get_xy()  # Get the position of the rectangle
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center', color='white', fontsize=10)

# Add title and labels
plt.title("Percentage of Churned Customers by Senior Citizen Status", fontsize=12)
plt.xlabel('Senior Citizen Status')
plt.ylabel('Percentage %')

# Show the plot
plt.tight_layout()
plt.show()
```

#A notably higher percentage of senior citizens have churned compared to non-senior citizens

```python
plt.figure(figsize = (9, 4))  # Resize

# Plot the histogram with custom color palette
sns.histplot(x = "tenure", data = df, kde = True, bins = 72, hue = "Churn", palette = custom_colors)

plt.show()
```

```python
# List of features to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the figure and axes for subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 15))  # 3x3 grid for 9 plots
axes = axes.flatten()  # Flatten to make it easier to loop over

# Loop over the columns and plot the count plot for each
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i], palette =custom_colors, hue = df["Churn"])
    axes[i].set_title(f'Count of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

This visual represents categorical counts of customers based on services they use, segmented by churn status (Yes/No). Key insights indicate that customers without internet-related services, such as online security, backup, and streaming, tend to churn less. Fiber optic internet users have higher churn rates, and features like tech support and device protection appear to influence retention positively. The data highlights significant relationships between service subscriptions and customer churn.
